```python

import json
from pathlib import Path

import matplotlib.pyplot as plt

RUN_DIR = Path("..") / "outputs" / "efficientnet_b0_head_20260505_151350"

history_path = RUN_DIR / "history.json"

print("History path:", history_path)
print("Exists:", history_path.exists())
```

```python
with open(history_path, "r", encoding="utf-8") as f:
    history = json.load(f)

epochs = [h["epoch"] for h in history]
train_loss = [h["train_loss"] for h in history]
val_loss   = [h["val_loss"] for h in history]
train_acc  = [h["train_acc"] for h in history]
val_acc    = [h["val_acc"] for h in history]

epochs
```

```python
plt.figure()
plt.plot(epochs, train_loss, marker="o", label="Train Loss")
plt.plot(epochs, val_loss, marker="o", label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Curves – ResNet50")
plt.legend()
plt.tight_layout()

out_loss = RUN_DIR / "loss_curves.png"
plt.savefig(out_loss, dpi=150)
plt.show()

print("Saved:", out_loss)
```

```python
plt.figure()
plt.plot(epochs, train_acc, marker="o", label="Train Accuracy")
plt.plot(epochs, val_acc, marker="o", label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Accuracy Curves – ResNet50")
plt.legend()
plt.tight_layout()

out_acc = RUN_DIR / "accuracy_curves.png"
plt.savefig(out_acc, dpi=150)
plt.show()

print("Saved:", out_acc)
```

```python
best = max(history, key=lambda x: x["val_acc"])
print("Best epoch:", best["epoch"])
print("Best validation accuracy:", best["val_acc"])
```
